```python
#Read data files. 
#MMG - Map the Meal Gap.
mmg_2019_2022 = pd.read_csv("MMG2024_2019-2022_Data_ToShare_v3.csv")
mmg_2019_2022.head(5)

```

```python
#Reduce data set to desired records by converting the 'State' field to string and selecting records with state = 'KY"
mmg_2019_2022 = mmg_2019_2022.loc[mmg_2019_2022.State == 'KY']
mmg_2019_2022.head(2)
```

```python
#verify that it has 480 and drop state fields since it is no longer needed
print(mmg_2019_2022.shape)
mmg_2019_2022.drop(columns='State', inplace=True)
```

Rename columns to have shorter or more descriptive names. I am Using the following acronyms: FIR = Food insecurity rate and FI = Food insecure with


```python
mmg_2019_2022.columns = mmg_2019_2022.columns.str.strip()
print(mmg_2019_2022.columns)
```

```python

mmg_2019_2022 = mmg_2019_2022.rename(columns={'County, State': 'COUNTY', 'Overall Food Insecurity Rate' :'FIR', '# of Food Insecure Persons Overall' :'FI_PERSONS'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'Food Insecurity Rate among Black Persons (all ethnicities)' : 'FIR_BLACK'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'Food Insecurity Rate among Hispanic Persons (any race)' :'FIR_HISPANIC'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'Food Insecurity Rate among White, non-Hispanic Persons': 'FIR_WHITE'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'SNAP Threshold': 'SNAP_THRESHOLD','% FI ≤ SNAP Threshold' :  'PER_FI_SNAP_ELIG', '% FI > SNAP Threshold' :  'PER_FI_NOT_SNAP_ELIG'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'Child Food Insecurity Rate' : 'FIR_CHILDREN', '# of Food Insecure Children': 'FI_CHILDREN'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'% food insecure children in HH w/ HH incomes below 185 FPL':'PER_FI_CHILDREN_WIC_ELIG'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'% food insecure children in HH w/ HH incomes above 185 FPL':'PER_FI_CHILDREN_NOT_WIC_ELIG'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'Cost Per Meal': 'COST_PER_MEAL_NON_FI', 'Weighted weekly $ needed by FI' : 'WEEKLY_MEAL_COST_F1'})
mmg_2019_2022 = mmg_2019_2022.rename(columns={'Weighted Annual Food Budget Shortfall' : 'ANNUAL_FOOD_BUDGET_SHORTFALL'})
```

```python
print(mmg_2019_2022.columns)
```

```python
#Check for null values
mmg_2019_2022.isnull().sum()
```

```python
#Drop food insecurity rate by race/ethnicity since there's a large percentage of missing values for blacks and hispanics
mmg_2019_2022.drop(columns=['FIR_BLACK','FIR_HISPANIC', 'FIR_WHITE'], inplace=True)
mmg_2019_2022.isnull().sum()
```

```python
#drop additional column that I will  not be use
mmg_2019_2022.drop(columns=['FIPS','SNAP_THRESHOLD', 'PER_FI_SNAP_ELIG', 'PER_FI_NOT_SNAP_ELIG','PER_FI_CHILDREN_NOT_WIC_ELIG', 'PER_FI_CHILDREN_WIC_ELIG'], inplace=True)
print(mmg_2019_2022.columns)
```

```python
mmg_2019_2022.dtypes
```

```python
# Change fields containg number to float, ecept those containg counts of people (those I will chnage to integers)
#Cnange county to string

mmg_2019_2022['COUNTY'] = mmg_2019_2022['COUNTY'].astype('string')
mmg_2019_2022['FIR'] = mmg_2019_2022['FIR'].str.replace('%', '').astype(float) / 100
mmg_2019_2022['FI_PERSONS'] = mmg_2019_2022['FI_PERSONS'].str.replace(',', '').astype(int)
mmg_2019_2022['FIR_CHILDREN'] = mmg_2019_2022['FIR_CHILDREN'].str.replace('%', '').astype(float) / 100
mmg_2019_2022['COST_PER_MEAL_NON_FI'] = mmg_2019_2022['COST_PER_MEAL_NON_FI'].str.replace('$', '').astype(float) / 100
mmg_2019_2022['WEEKLY_MEAL_COST_F1'] = mmg_2019_2022['WEEKLY_MEAL_COST_F1'].str.replace('$', '').astype(float) / 100
#mmg_2019_2022['ANNUAL_FOOD_BUDGET_SHORTFALL'] = mmg_2019_2022['ANNUAL_FOOD_BUDGET_SHORTFALL'].str.replace('$', '').replace(',', '').strip.astype(float) / 100
mmg_2019_2022['ANNUAL_FOOD_BUDGET_SHORTFALL'] = pd.to_numeric(mmg_2019_2022['ANNUAL_FOOD_BUDGET_SHORTFALL'].str.replace(r'[\$,%]', '', regex=True))
mmg_2019_2022['FI_CHILDREN'] = mmg_2019_2022['FI_CHILDREN'].str.replace(',', '').str.replace('%', '').str.replace('-', '0').astype(int)
mmg_2019_2022.dtypes
```

```python
#Check for null values
mmg_2019_2022.isnull().sum()
```

```python
#Per data % 'Any estimated food insecurity rates of 0% should not be interpreted literally (e.g., 0% among white, non-Hispanic persons in the District of Columbia in 2020'
#Therefore, I will replace NAN values with -1
mmg_2019_2022 = mmg_2019_2022.fillna(-1)
mmg_2019_2022.isnull().sum()
```

```python
mmg_2019_2022.tail(2)
```

```python
#Creating regions for better graphics. But the grouping is not ideal due to counties with overlaping regions.
mmg_2019_2022['REGIONS'] = 'KENCTUCKY GEOGRAPHICAL REGIONS'
mmg_2019_2022['COUNTY'] = mmg_2019_2022['COUNTY'].str.replace(' County, Kentucky', '', regex=False)
mmg_2019_2022.loc[mmg_2019_2022['COUNTY'].isin(['Anderson','Bourbon', 'Boyle', 'Clark', 'Fayette', 'Franklin', 'Garrard', 'Harrison', 'Jessamine', 'Lincoln', 'Madison', 'Mercer', 'Montgomery', 'Nicholas', 'Scott', 'Woodford']), 'REGIONS'] = 'Bluegrass'
mmg_2019_2022.loc[mmg_2019_2022['COUNTY'].isin(['Ballard','Caldwell','Christian','Crittenden','Daviess','Graves','Hancock','Henderson','Hopkins','Jackson','Jefferson','Jones','McLean','Muhlenberg','Ohio','Union','Webster',]), 'REGIONS'] = 'Western Coal Fields'
mmg_2019_2022.loc[mmg_2019_2022['COUNTY'].isin(['Bell', 'Boyd', 'Carter', 'Elliot','Greenup', 'Johnson', 'Knott', 'Knox', 'Laurel', 'Lawrence', 'Lee', 'Leslie', 'Letcher', 'Magoffin', 'Martin',  'McCreary', 'Morgan', 'Owsley', 'Perry', 'Pike', 'Whitley','Wolfe']), 'REGIONS'] = 'Knobs'
mmg_2019_2022.loc[mmg_2019_2022['COUNTY'].isin(['Breathitt','Clay','Estill','Floyd','Harlan','Jackson','Johnson','Knott','Laurel','Lee','Leslie','Letcher','Lewis','Magoffin','Martin','Menifee','Morgan','Owsley','Perry','Pike','Powell','Rockcastle','Rowan','Whitley']), 'REGIONS'] = 'Eastern Coal Fields'
mmg_2019_2022.loc[mmg_2019_2022['COUNTY'].isin(['Ballard', 'Calloway', 'Carlisle', 'Fulton', 'Graves', 'Hickman', 'Jackson', 'Jefferson', 'Lyon', 'Marshall', 'McCracken', 'Murray']), 'REGIONS'] = 'Jackson Purchase'
mmg_2019_2022.loc[mmg_2019_2022['REGIONS'] == 'KENCTUCKY GEOGRAPHICAL REGIONS', 'REGIONS'] = 'Pennyrile'
mmg_2019_2022.head(10) 
```

```python
#pd.set_option('display.max_rows', None)  
SUM_FI = mmg_2019_2022.groupby(['Year','REGIONS']).agg({'FI_PERSONS': 'sum'}).reset_index()

print(SUM_FI)
#All Average FIR by county are below 30%
```

```python
#Prepare data for graphing
SUM_FI_PIVOTED = SUM_FI.pivot(index='Year', columns='REGIONS', values='FI_PERSONS')
SUM_FI_PIVOTED
```

```python
#Create a nulti-line graph

# Create the plot
plt.figure(figsize=(20, 12)) 

SUM_FI_PIVOTED.plot(kind='line', marker='o')

# Add title and labels
plt.title('Food Insecured Population by Region and Year')
plt.xlabel('Year')
plt.ylabel('FI_PERSONS')

# Show the plot
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

Though the regional grouping was challinging and is not 100% accurate, the graph shows no drasticc difference between regions.
